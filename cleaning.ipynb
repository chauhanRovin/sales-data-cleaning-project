```python
import pandas as pd

# Load dataset
df = pd.read_csv('dataset.csv')

print("Original Data:")
print(df)

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())

# Clean column names
df.columns = df.columns.str.strip()

print("\nCleaned Data:")
print(df)

# Save cleaned file
df.to_csv('cleaned_data.csv', index=False)
```
